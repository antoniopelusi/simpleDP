```python
import matplotlib.pyplot as plt
import random
```

```python
#list containing the possible values
rawSet = [0, 1]

#list containing multiple rawSet: more bits == more accuracy
rawData = []

for i in range(10000): #range(10000) == 10000x 1 bits and 10000 0 bits
    rawData.extend(rawSet)

#function that plot the original dataset stats
def plot_no_DP():
  plotData = {'0': rawData.count(0), '1': rawData.count(1)}
  labels = list(plotData.keys())
  values = list(plotData.values())

  graph = plt.bar(labels, values)
  graph[0].set_color('darkred')
  graph[1].set_color('darkgreen')

  for bar in graph:
    plt.annotate(bar.get_height(), xy=(bar.get_x()+0.33, bar.get_height()-1000), fontsize=10, color="white")
    
  plt.show()

#function that plot the dataset stats after the DP algorithm
def plot_with_DP():
  dpData = []

  for bit in rawData:
    flip = random.randint(0,1)
    if flip == 0:
      dpData.append(bit)
    else:
      dpData.append(random.randint(0,1))

  dpPlotData = {'0': dpData.count(0), '1': dpData.count(1)}

  labels = list(dpPlotData.keys())
  values = list(dpPlotData.values())

  graph = plt.bar(labels, values)
  graph[0].set_color('darkred')
  graph[1].set_color('darkgreen')

  for bar in graph:
    plt.annotate(bar.get_height(), xy=(bar.get_x()+0.33, bar.get_height()-1000), fontsize=10, color="white")

  plt.show()
```

```python
#plot original dataset stats
plot_no_DP()
```

```python
#add a new bit
rawData.append(1)
```

```python
#plot again the stats, the new bit is identifiable (we know that it is a 1)
plot_no_DP()
```

```python
#plot the stats after DP algorithm
plot_with_DP()
```

```python
#add a new bit
rawData.append(1)
```

```python
#plot again the stats, the new stats are not 100% accurate but the new bits is no longer identifiable (we don't know if it is a 0 or 1)
plot_with_DP()
```
